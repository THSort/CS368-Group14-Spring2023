```python
from bs4 import BeautifulSoup
import os
import re
import html
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
papa_list = []
```

```python
# HUMAN ASSIGNMENT 1

files = [f for f in os.listdir('./Research_Metrics/Human/Assignment 1/Task 1') if f.endswith('.xml')]

created_flag = 0 
df = pd.DataFrame(columns=['vg_b', 'vg_e', 'vg_b_max', 'vg_b_avg', 'vg_e_max', 'vg_e_avg', 'params', 'locphy', 'locpro', 'locbl', 'loccom', 'n', 'n1', 'n2', 'n', 'n1', 'n2', 'v', 'b', 'd', 'e', 'l', 'maxnd', 'miwoc', 'micwc', 'mi'])

for _file in sorted(files):
    filename = "./Research_Metrics/Human/Assignment 1/Task 1/" + _file
    with open(filename, 'r') as f:
        file_content = f.read() 

    soup = BeautifulSoup(file_content, 'html')
    children = soup.findChildren('file_total', recursive=True)[0]
    children = str(children)

    temp = re.split('<operators>', children)[0]
    temp = temp + "</file_total>"
    soup = BeautifulSoup(temp, 'html')

    root_tag = soup.find()
    metrics_tags = root_tag.find_all(recursive=False)

    names = []
    values = []

    for tag in metrics_tags:
        metric_name = ""
        tag_name = str(tag)
        for char in tag_name:
            if(char == "<"):
                continue
            if(char == ">"):
                break

            metric_name = metric_name + char 
        
        metric_value = tag.text 
        if(metric_name != 't'):
            names.append(metric_name.split(' ')[0])
            values.append(float(metric_value))

    df.loc[len(df)] = values

col_means = df.mean()
papa_list.append(list(col_means))
```
